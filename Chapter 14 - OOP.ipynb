1. Write a class called Investment with fields called principal and interest. The constructor should set the values of those fields. There should be a method called value_after that
returns the value of the investment after n years. The formula for this is $p(1 + i)^n$, where p is
the principal, and i is the interest rate. It should also use the special method __str__ so that
printing the object will result in something like below $\newline$
Principal - $1000.00, Interest rate - 5.12%

```python
class Investment:
    def __init__(self, p, i, n):
        self.p = p
        self.i = i
        self.n = n
        
    def value_after(self):
        self.value = round(self.p*((1 + self.i/100)**self.n), 4)
    
    def print_value(self):
        self.value_after()
        print(f"Principal - ${self.p}, Interest rate - {self.i}%, Years - {self.n} = Compound Interest: {self.value}")
        
principal = eval(input("Enter pricipal: "))
interest = eval(input("Enter interest rate: "))
years = eval(input("Enter years: "))

investment = Investment(principal, interest, years)
investment.print_value()
```

10 and 99 items, and a 20% discount is applied for orders of 100 or more items. There should

```python
class Product:
    def __init__(self, name, amount, price):
        name = name.strip().lower().capitalize()
        self.name = name
        self.amount = amount
        self.price = price
        
    def get_price(self):
        if 0 <= self.amount < 10:
            cost_per_item = self.price
        elif 10 <= self.amount < 100:
            cost_per_item = self.price*(0.9)
        else:
            cost_per_item = self.price*(0.8)   
        return (round(cost_per_item, 4))
    
    def make_purchase(self):
        total_cost = self.amount*self.get_price()
        return(total_cost)
        
n = input("Enter the name of the product: ")
s = eval(input("Enter how many items do you want to buy: "))
p = eval(input("Enter the price of the item: $"))

product = Product(n, s, p)
print(f"\nProduct name: {product.name}")
print(f"Amout of items: {product.amount}")
print(f"Original Price of item: ${product.price}")
print(f"Total cost with discount: ${product.make_purchase()}")
```

4. Write a class called Time whose only field is a time in seconds. It should have a method called
convert_to_minutes that returns a string of minutes and seconds formatted as in the following example: if seconds is 230, the method should return '5:50'. It should also have
a method called convert_to_hours that returns a string of hours, minutes, and seconds
formatted analogously to the previous method.

```python
class Time:
    def __init__(self, seconds):
        self.seconds = seconds
    
    def convert_to_minutes(self):
        minutes = self.seconds//60
        sec = (self.seconds - 60*minutes)%60
        if sec < 10:
            return f"{minutes}:0{sec}"
        else:
            return f"{minutes}:{sec}"
        
    def convert_to_hour(self):
        hour = self.seconds//(60*60)
        minutes = (self.seconds - (60*60)*hour)//60
        sec = self.seconds%60
        if minutes < 10:
            if sec < 10:
                return f"{hour}:0{minutes}:0{sec}"
            else:
                return f"{hour}:0{minutes}:{sec}"
        else:
            if sec < 10:
                return f"{hour}:{minutes}:0{sec}"
            else:
                return f"{hour}:{minutes}:{sec}"

s = eval(input("Enter how many seconds: "))

time = Time(s)
print(f"{s} seconds to minute: {time.convert_to_minutes()}")
print(f"{s} seconds to hour: {time.convert_to_hour()}")
```

5. Write a class called Wordplay. It should have a field that holds a list of words. The user
of the class should pass the list of words they want to use to the class. There should be the
following methods: $\newline$
• words_with_length(length) — returns a list of all the words of length length$\newline$
• starts_with(s) — returns a list of all the words that start with s$\newline$
• ends_with(s) — returns a list of all the words that end with s$\newline$
• palindromes() — returns a list of all the palindromes in the list$\newline$
• only(L) — returns a list of the words that contain only those letters in L$\newline$
• avoids(L) — returns a list of the words that contain none of the letters in L

```python
class Wordplay:
    def __init__(self, L):
        self.L = L
    
    def words_with_length(self):
        length = eval(input("Enter length of a string: "))
        lw = [word for word in self.L if len(word) == length]
        return(lw)
    
    def start_with(self):
        letter = input("Enter a letter: ")
        lw = [word for word in self.L if word[0] == letter]
        return(lw)
    
    def ends_with(self):
        letter = input("Enter a letter: ")
        lw = [word for word in self.L if word[-1] == letter]
        return(lw)
    
    def palindromes(self):
        lw = [word for word in self.L if word == word[::-1]]
        return(lw)
    
    def only(self):
        lw = list()
        list_of_letters = eval(input("Enter a list of letters: "))
        list_of_letters = list(dict.fromkeys(list_of_letters))
        for word in self.L:
            count = 0
            for letter in list_of_letters:
                count += word.count(letter)
            
            if count == len(word):
                lw.append(word)
        
        return(lw)
    
    def avoids(self):
        lw = list()
        list_of_letters = eval(input("Enter a list of letters: "))
        list_of_letters = list(dict.fromkeys(list_of_letters))
        for word in self.L:
            count = 0
            for letter in list_of_letters:
                count += word.count(letter)
            
            if count == 0:
                lw.append(word)
        return(lw)
    
list_word = eval(input("Enter a list of words: "))
wordplay = Wordplay(list_word)
wordplay.words_with_length()
wordplay.start_with()
wordplay.ends_with()
wordplay.palindromes()
wordplay.only()
wordplay.avoids()
```

7. Use the Standard_deck class of this section to create a simplified version of the game War.
In this game, there are two players. Each starts with half of a deck. The players each deal
the top card from their decks and whoever has the higher card wins the other player’s cards
and adds them to the bottom of his deck. If there is a tie, the two cards are eliminated from
play (this differs from the actual game, but is simpler to program). The game ends when one
player runs out of cards.

```python
import random
class Standard_deck():
    def __init__(self):
        self.cards = []
        for s in ['Hearts', 'Diamonds', 'Clubs', 'Spades']:
            for v in range(1,14):
                self.cards.append((s, v))
        
    def shuffle(self):
        random.shuffle(self.cards)
        self.player1 = self.cards[:26]
        self.player2 = self.cards[26:]
    
    def rounds(self):
        card1 = self.player1[0]
        card2 = self.player2[0]

        if card1[1] > card2[1]:
            self.player2.remove(card2)
            self.player1.append(card2)
        elif card1[1] < card2[1]:
            self.player1.remove(card1)
            self.player2.append(card1)
        else:
            self.player1.remove(card1)
            self.player2.remove(card2)
            
    def check(self):
        if len(self.player2) == 0:
            print("\nPlayer 1 Wins!")
        elif len(self.player1) == 0:
            print("\nPlayer 2 Wins!")
            
deck = Standard_deck()
deck.shuffle()
rounds = 1
while True:
    print(f"\nRound {rounds}")
    print(f"\nPlayer 1: {deck.player1}")
    print(f"\nPlayer 2: {deck.player2}")
    if len(deck.player1) == 0 or len(deck.player2) == 0:
        break
    else:
        deck.rounds()
        rounds += 1
deck.check()
```

9. Write a class called Rock_paper_scissors that implements the logic of the game Rockpaper-scissors. For this game the user plays against the computer for a certain number of
rounds. Your class should have fields for the how many rounds there will be, the current
round number, and the number of wins each player has. There should be methods for getting
the computer’s choice, finding the winner of a round, and checking to see if someone has one
the (entire) game. You may want more methods.

```python
from math import ceil

class Rock_paper_scissors:
    def __init__(self, n_rounds):
        self.n_rounds = n_rounds
        self.c_round = 0
        self.n_wins1 = 0
        self.n_wins2 = 0
        
    def game(self):
        print(f"\nRound {self.c_round}")
        choice = ["Rock", "Paper", "Scissors"]
        while True:
            pick1 = input("Enter Rock, Paper or Scissors: ")
            pick1 = pick1.strip().lower().capitalize()
            if pick1 in choice:
                break
            else:
                print("Invalid Input!\n")
        pick2 = random.choice(choice)
        
        print(f"Player 1: {pick1}")
        print(f"Player 2: {pick2}")
        
        if pick1 == "Rock" and pick2 == "Scissors":
            print("Win.")
            self.n_wins1 += 1
        elif pick1 == "Rock" and pick2 == "Paper":
            print("Lose.")
            self.n_wins2 += 1
        elif pick1 == "Paper" and pick2 == "Rock":
            print("Win.")
            self.n_wins1 += 1
        elif pick1 == "Paper" and pick2 == "Scissors":
            print("Lose.")
            self.n_wins2 += 1
        elif pick1 == "Scissors" and pick2 == "Paper":
            print("Win.")
            self.n_wins1 += 1
        elif pick1 == "Scissors" and pick2 == "Rock":
            print("Lose.")
            self.n_wins2 += 1
        else:
            print("Tie")
            
        print(f"Wins: {self.n_wins1}")
        
    def rounds(self):
        self.c_round += 1
        
    def check(self):
        if self.n_wins1 == ceil(self.n_rounds/2):
            return 1
        elif self.n_wins2 == ceil(self.n_rounds/2):
            return 2
        else:
            return 0
        
RPC = Rock_paper_scissors(5)
while RPC.c_round < 5:
    RPC.rounds()
    RPC.game()
    
    if RPC.c_round < 5:
        if RPC.check() == 1:
            print("\nPlayer 1 Wins.")
            break
        elif RPC.check() == 2:
            print("\nPlayer 2 Wins.")
            break
    else:
        if RPC.check() == 1:
            print("\nPlayer 1 Wins.")
            break
        elif RPC.check() == 2:
            print("\nPlayer 2 Wins.")
            break
        else:
            print("\nTie.")
```
